```python
from google.colab import files
uploaded = files.upload()
```

```python
import PyPDF2

# 1. Get the exact name of the file you just uploaded
file_name = list(uploaded.keys())[0]

# 2. Tell PyPDF2 to open that specific file
reader = PyPDF2.PdfReader(file_name)

# 3. Tell it to go to the very first page (Page 0)
first_page = reader.pages[0]

# 4. Extract the text and save it
extracted_text = first_page.extract_text()

# 5. Show us the text!
print(extracted_text)
```
